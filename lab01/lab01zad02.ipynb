A: Zaimportowanie potrzebnych paczek, wyświetlenie tabeli *miasta* oraz czystych wartości tabeli *miasta.values*.

```python
import pandas as pd
import matplotlib.pyplot as plt
miasta = pd.read_csv("miasta.csv")
print(miasta)
print("\n---------------------------------\n")
print(miasta.values)
```

B: Dodawanie do tabeli wiersza z 2010 rokiem.

```python
new_row = pd.DataFrame({'Rok': [2010], 'Gdansk': [460], 'Poznan': 555, 'Szczecin': 405})
miasta.loc[10] = new_row.iloc[0]
print(miasta)
```

C: Wykres dla ludności Gdańska.

```python
gdansk = miasta['Gdansk']
rok = miasta['Rok']

plt.figure(figsize=(10, 6))
plt.plot(rok, gdansk, 'ro-')
plt.title('Ludność w miastach Polski')
plt.xlabel('Lata')
plt.ylabel('Liczba ludności [w tys]')

plt.show()
```

D: Wykres ze zmianami ludności we wszystkich miastach z tabeli.

```python
poznan = miasta['Poznan']
szczecin = miasta['Szczecin']
plt.figure(figsize=(10, 6))
plt.plot(rok, gdansk, label="Gdańsk")
plt.plot(rok, poznan,  label="Poznan")
plt.plot(rok, szczecin, label="Szczecin")
plt.legend(["Gdańsk", "Poznań", "Szczecin"])
plt.title('Ludność w miastach Polski')
plt.xlabel('Lata')
plt.ylabel('Liczba ludności [w tys]')
plt.show()
```

E: Standaryzacja danych przy użyciu paczki numpy.

```python
import numpy as np
all_values = miasta.values.flatten()
for i in all_values:
    if i >= 1900:
        all_values = np.delete(all_values, np.where(all_values == i))

mean_all_values = np.mean(all_values)
sd_all_values = np.std(all_values, ddof=1)
z = (all_values - mean_all_values) / sd_all_values


print("Wszystkie dane w tablicy:",  all_values)
print("Przeskalowane dane:", z)
print("Średnia po skalowaniu:", round(np.mean(z), 2))
print("Odchylenie standardowe po skalowaniu:", np.std(z, ddof=1))
```

Średnia po skalowaniu wynosi 0, natomiast odchylenie standardowe jest równe 1.

F: Normalizacja danych przy użyciu paczki numpy.

```python
min = np.min(all_values)
max = np.max(all_values)

z = (all_values - min) / (max - min)

print("Wszystkie dane w tablicy:",  all_values)
print("Przeskalowane dane:", z)
print("Minimalna wartość po skalowaniu:", np.min(z))
print("Maksymalna wartość po skalowaniu:", np.max(z))

```

Minimalna wartość po skalowaniu wynosi 0, a maksymalna to 1.
