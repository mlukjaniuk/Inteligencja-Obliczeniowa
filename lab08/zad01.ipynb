```python
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
with open ('article.txt', 'r') as file:
    text = file.read()
    tokens = nltk.tokenize.word_tokenize(text)
    print(tokens)
    print("Liczba słów po tokenizacji: ", len(tokens))
```

```python
with open ('article.txt', 'r') as file:
    text = file.read()
    text = text.lower()
    tokens = nltk.tokenize.word_tokenize(text)
    stop_words = set(nltk.corpus.stopwords.words('english'))
    filtered_tokens = [w for w in tokens if not w in stop_words]
    print(filtered_tokens)
    print("Liczba słów po usunięciu stopwords: ", len(filtered_tokens))
```

```python
gist_file = open("gist_stopwords.txt", "r")
old_stopwords = nltk.corpus.stopwords.words('english')
try:
    content = gist_file.read()
    new_stopwords = content.split(",")
    new_stopwords.append("\'s")
    new_stopwords.append("\'ve")
    new_stopwords.append("n\'t")
    new_stopwords.append("\'")
    new_stopwords.append(".")
    new_stopwords.append(",")
    new_stopwords.append(":")
    new_stopwords.append("``")
    new_stopwords.append("''")
    new_stopwords.append("-")
    new_stopwords.append("[")
    new_stopwords.append("]")

    for word in new_stopwords:
        if word not in old_stopwords:
            old_stopwords.append(word)
finally:
    gist_file.close()

filtered_tokens = [w for w in tokens if not w in new_stopwords]
print(filtered_tokens)
print("Liczba słów po usunięciu nowych stopwords: ", len(filtered_tokens))
```

```python
# WordNet Lemmatizer
lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_tokens]
print(lemmatized_words)
print("Liczba słów po lematyzacji: ", len(lemmatized_words))
```

```python
word_freq = nltk.FreqDist(lemmatized_words)
top_words = word_freq.most_common(10)
words = [word[0] for word in top_words]
freqs = [word[1] for word in top_words]
plt.figure(figsize=(10, 6))
plt.bar(words, freqs)
plt.xlabel('Słowa')
plt.ylabel('Liczba wystąpień')
plt.title('10 najczęściej występujących słów')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
word_string = ' '.join(lemmatized_words)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(word_string)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
